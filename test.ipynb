```python
from functools import wraps

current_user = {
    "role": "admin",
    "is_authenticated": True
}


def role_required(role):  # deco factory
  def wrapper(func):  # decorator (func -> decorated func)
    @wraps(func)
    def decorated_func(*args, **kwargs):
      if current_user['is_authenticated'] and current_user["role"] == role:
        return func(*args, **kwargs)

    return decorated_func
  return wrapper
```

```python
@role_required("admin")
def private(name):
  return f"hello {name}, here are my secrets"


@role_required("manager")
def passwords():
  return "very very secret passwords"


private("harikesh")
passwords()

# smart_private = smart_maker(private, "admin")
# smart_passwords = smart_maker(passwords, "manager")

# smart_private()
# smart_passwords()
```
